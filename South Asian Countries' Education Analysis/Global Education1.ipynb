## South Asian Countries Education Analysis

```python
# Making a new dataframe of south asian countries
south_asian_countries = ['Afghanistan', 'Bangladesh', 'Bhutan', 'India', 'Maldives', 'Nepal', 'Pakistan', 'Sri Lanka']
south_asian_countries = df[df['Country'].isin(south_asian_countries)]
south_asian_countries.head(10) 
```

```python
# Resetting the index of the new dataframe
south_asian_countries.reset_index(inplace=True, drop=True)
south_asian_countries.head(10)
```

### Dealing with zero values
There are many columns with zero values, so for better analysis let's replace them with the mean of their respective columns.

```python
# Calculating the mean of each column
mean_values = south_asian_countries.iloc[:,3:-1].mean()

# Replacing zero values with the mean of their respective columns
south_asian_countries = south_asian_countries.replace(0, mean_values)
south_asian_countries.head(10)
```

### Proficiency of Grade 2 & 3 students in Reading and Maths

```python
# Selecting the required columns
proficiency = south_asian_countries[['Country', 'Grade_2_3_Proficiency_Reading', 'Grade_2_3_Proficiency_Math']]

# Setting the index as the country names
proficiency.set_index('Country', inplace=True)

# Plotting the data
proficiency.plot(kind='bar', figsize=(12, 6))
plt.title('Grade 2-3 Proficiency Reading and Math in South Asian Countries')
plt.xlabel('Country')
plt.ylabel('Proficiency Score')
plt.legend(['Reading', 'Math'])

# Rotating x-axis labels for better visibility
plt.xticks(rotation=45)
plt.show()
```

Reading seems to be more popular than Math among students, especially in countries like Bangladesh and Pakistan, where the difference is significant. In contrast, in countries like India and Afghanistan, Math appears to be students' favorite subject.

### OOSR for Lower Secondart Education

```python

# Extracting the required columns
female_data = south_asian_countries['OOSR_Lower_Secondary_Age_Female']
male_data = south_asian_countries['OOSR_Lower_Secondary_Age_Male']

# Creating the plot
plt.figure(figsize=(10, 6))
plt.bar(south_asian_countries['Country'], female_data, label='Female')
plt.bar(south_asian_countries['Country'], male_data, bottom=female_data, label='Male')
plt.xlabel('Country')
plt.ylabel('Count')
plt.title('OOSR of Lower Secondary age Females and Males in South Asian Countries')
plt.xticks(rotation=90)
plt.legend()

# Show the plot
plt.show()
```

In all South Asian countries, there is a significant gap between the out-of-school rates for males and females, with the rate being much lower for females. This disparity highlights a troubling reality about our society.

### Literacy Rate among youths

```python
# Selecting the required columns
literacy_rate = south_asian_countries[['Country', 'Youth_15_24_Literacy_Rate_Male', 'Youth_15_24_Literacy_Rate_Female']]

# Setting the index as the country names
literacy_rate.set_index('Country', inplace=True)

# Plotting the data
literacy_rate.plot(kind='bar', figsize=(12, 6))
plt.title('Youth Literay Rate in South Asian Countries')
plt.xlabel('Country')
plt.ylabel('Literacy Rate')
plt.legend(['Youth_15_24_Literacy_Rate_Male', 'Youth_15_24_Literacy_Rate_Female'])
plt.xticks(rotation=45)
plt.show()
```

Despite the significant gap in the out-of-school rate (OOSR) between males and females in the lower secondary age group, the literacy rate among them is quite comparable.

### Unemployment Rate

```python
# Selecting the required columns
unemployed = south_asian_countries[['Unemployment_Rate','Country']]

#setting country as index
unemployed = unemployed.set_index('Country')

# Plotting the data of unemployment
unemployed.plot(kind='bar', figsize=(12, 6))
plt.title('Unemployment Rate')
plt.xlabel('Country')
plt.ylabel('Unemployment Rate (%)')
plt.xticks(rotation=45)
plt.show()
```

### Literacy VS Unemployment

```python
# Taking the mean of the Literacy Rate of Males and Female
south_asian_countries['Literacy_Rate'] = south_asian_countries[['Youth_15_24_Literacy_Rate_Male', 'Youth_15_24_Literacy_Rate_Female']].mean(axis=1)
south_asian_countries.head()
```

```python
# Selecting the required columns
comparison = south_asian_countries[['Country','Literacy_Rate','Unemployment_Rate']]

# Setting the index as the country names
comparison.set_index('Country', inplace=True)

# Plotting the data
comparison.plot(kind='bar', figsize=(12, 6))
plt.title('Literacy VS Unemployment')
plt.xlabel('Country')
plt.ylabel('Literacy Rate')
plt.legend(['Literacy_Rate', 'Unemployment Rate'])
plt.xticks(rotation=45)
plt.show()
```
